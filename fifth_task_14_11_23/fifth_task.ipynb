### 1. Scipy - обрезка изображения

```python
from scipy import datasets
import matplotlib.pyplot as plt
import numpy as np

face = datasets.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
```

```python
crop_face = face[100:-100, 100:-100]
plt.imshow(crop_face, cmap=plt.cm.gray)
```

```python
sy, sx = face.shape
y, x = np.ogrid[0:sy, 0:sx]
y.shape, x.shape

centerx, centery = (660, 300)
mask = ((y - centery)**2 + (x - centerx)**2) > 230**2
face[mask] = 0

plt.imshow(face, cmap=plt.cm.gray)
```

### 2. Работа со статистикой

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data/populations.txt')
years, hares, lynxes, carrots = data.T

plt.axes([0.2, 0.1, 0.5, 0.8])
plt.plot(years, hares, years, lynxes, years, carrots)
plt.legend(('Hare', 'Lynx', 'Carrot'), loc=(1.05, 0.5))
plt.show()
```

```python
print(f'Mean for hares:  {np.mean(hares)}, standart deviation: {np.std(hares)}')
print(f'Mean for lynxes: {np.mean(lynxes)}, standart deviation: {np.std(lynxes)}')
print(f'Mean for carrots: {np.mean(carrots)}, standart deviation: {np.std(carrots)}')
```

```python
print(f'Max population of hares: {years[np.argmax(hares)]}')
print(f'Max population of lynxes: {years[np.argmax(lynxes)]}')
print(f'Max population of carrots: {years[np.argmax(carrots)]}')
```
